```python
# Plot the correlation
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df[['age','avg_salary','Rating','desc_len','num_comp']].corr(),vmax=.3, center=0, cmap=cmap,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})


# Generate correlation matrix
corr_matrix = df[['age', 'avg_salary', 'Rating', 'desc_len', 'num_comp']].corr()

# Define colorscale
colorscale = [[0, '#FFFFFF'], [1, '#FF0000']]  # Example colorscale

# Create heatmap figure
fig = go.Figure(data=go.Heatmap(
                   z=corr_matrix.values,
                   x=corr_matrix.columns,
                   y=corr_matrix.columns,
                   colorscale=colorscale,
                   hoverongaps = False))

# Update layout
fig.update_layout(title='Correlation Heatmap',
                  )

# Show plot
fig.show()
```

```python
# To visulaize the categorical data

# Selecting categorical columns
df_cat = df[['Location', 'Headquarters', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'company_txt', 
             'job_state', 'same_state', 'python_yn', 'R_yn', 'sql', 'aws', 'excel', 'powerbi_tableau', 'job_simp', 'seniority']]

# Loop through each column and create a horizontal bar chart
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts().reset_index() # Reset index to convert Series to DataFrame
    cat_num.columns = [i, 'Count'] # Rename columns for clarity
    cat_num_sorted = cat_num.sort_values(by='Count', ascending=False) # Sort values by count
    print("graph for %s column: total = %d" % (i, len(cat_num))) # Print total unique values for the column
    fig = px.bar(cat_num_sorted, x=i, y='Count', color= i)
    fig.update_layout(showlegend=False)
    fig.update_xaxes(tickangle=45) # Rotate x-axis labels for better readability
    fig.show()
```

```python
# To visualize the top 10 location, headquarters and company by No of Job Listings

for i in df_cat[['Location','Headquarters','company_txt']].columns:
    cat_num = df_cat[i].value_counts()[:10]
    #cat_num.columns = [i, 'Count']
    print("graph for %s: total = %d" % (i, len(cat_num)))
    fig = px.bar(cat_num,x=cat_num.index, y=cat_num, color= cat_num.index)
    fig.update_traces(textposition='outside')
    fig.update_layout(
    xaxis=dict(title=None),
    yaxis=dict(title=None),
    showlegend=False)
    fig.update_xaxes(tickangle=-45) # Rotate x-axis labels for better readability
    fig.show()


    
    
```

```python
# To visualize the pivot table
pivot_table = pd.pivot_table(df, index=['job_simp', 'seniority'], values='avg_salary').reset_index()

# Create bar chart figure
fig = go.Figure()

for level in pivot_table['seniority'].unique():
    job_data = pivot_table[pivot_table['seniority'] == level]
    fig.add_trace(go.Bar(
        x=job_data['job_simp'],
        y=job_data['avg_salary'],
        name=level
    ))

# Update layout
fig.update_layout(
    title='Average Salary by Job title and Job level',
    xaxis_title='Job Title',
    yaxis_title='Average Salary',
    barmode= 'stack'
)

# Show plot
fig.show()
```

```python
# To create a world cloud to display the most used words from the Job description

from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


# Concatenate all job descriptions into a single string
words = " ".join(df['Job Description'])

def punctuation_stop(text):
    """Remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered

words_filtered = punctuation_stop(words)
text = " ".join([ele for ele in words_filtered])

# Generate WordCloud
wc = WordCloud(background_color="white", random_state=1, stopwords=STOPWORDS, max_words=2000, width=1500, height=800)
wc.generate(text)

# Convert WordCloud image to array
image = wc.to_array()

# Create a plotly express image
fig = px.imshow(image)

# Update layout
fig.update_layout(width=1500, height=800)
fig.update_layout(
    xaxis=dict(title=None, showticklabels=False),
    yaxis=dict(title=None, showticklabels=False
               ))

# Show plot
fig.show()

```
