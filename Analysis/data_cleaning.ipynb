```python
# Import necessary Libraries
import pandas as pd
```

```python
# Read the scraped data
df=pd.read_csv('glassdoor_jobs.csv')
df
```

```python
# Let's begin the data cleaning of thr data
# To clean the Salary Estimate column
# create a new column to show which row has Salary in per hour
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)

df = df[df['Salary Estimate'] != '-1']
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0]) # to remove glassdoor text
minus_Kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
# to remove additional text
min_hr = minus_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
# Add average salary column
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary+df.max_salary)/2

```

```python
# To convert hourly wage to annual 
df['min_salary'] = df.apply(lambda x: x.min_salary*2 if x.hourly ==1 else x.min_salary, axis =1)
df['max_salary'] = df.apply(lambda x: x.max_salary*2 if x.hourly ==1 else x.max_salary, axis =1)

df[df.hourly ==1][['hourly','min_salary','max_salary', 'avg_salary']]
```

```python
# To append the new salaries into the Avg_salary column
df['avg_salary']= (df.min_salary+df.max_salary)/2

df[df.hourly ==1][['hourly','min_salary','max_salary', 'avg_salary']]
```

```python
#To remove the rating from the Company name column
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3], axis = 1)

```

```python
#Extract the state the company that listed the job is located from the state field 
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.job_state.value_counts()
# to create a coulmn showing if the Job location is in the same state as the headquarters
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
```

```python
#Create a new column to calculate the Age of the company that listed the job
df['age'] = df.Founded.apply(lambda x: x if x <1 else 2024 - x)
```

```python
#parsing of job description for data science skills required (python, etc.)

#python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
 
#r studio 
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()

#sql 
df['sql'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df.sql.value_counts()

#aws or azure
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() or 'azure' in x.lower() else 0)
df.aws.value_counts()

#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()

#power bi or tableau
df['powerbi_tableau'] = df['Job Description'].apply(lambda x: 1 if 'power bi' in x.lower() or 'tableau' in x.lower() else 0)
df.aws.value_counts()
```

```python
# Create a new column to simplifier the Job title and to recorgnise job roles that are not data inclined

def title_simplifier(title):
    if 'data scientist' in title.lower() or 'science' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower()  or 'data architect' in title.lower()  or 'data modeler' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower() or 'analytics' in title.lower() or 'data analysis' in title.lower():
        return 'data analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower() or 'management' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    

# To create the new title column
df['job_simp'] = df['Job Title'].apply(title_simplifier)

# To display value count
df.job_simp.value_counts()
```

```python
# It seems the data consist of some rows that are not data inclines. let us see them which will be filtered out later
df[df.job_simp =='na'][['job_simp','Job Title','Job Description']]
```

```python
# To parse the Level job role (either senior or junior)) required from the data set
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
    
df['seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```

```python
# To Fix Los Angeles state
df['job_state']= df.job_state.apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df.job_state.value_counts()
```

```python
#  Job description length  to check how it affects the salary range which will used for further eda
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
df['desc_len']
```

```python
#Create a Competitor count column. This column is to identify the no of companies that are compititors to the company listing the job
df['num_comp'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x != '-1' else 0)
df['num_comp']
df['Competitors']
```

```python
# To further Clean the company column
df['company_txt'] = df.company_txt.apply(lambda x: x.replace('\n', ''))

df['company_txt']
```

```python

#Filter out 'na' values in the 'job_simp' column to only display data inclined roles
df = df[df['job_simp'] != 'na'] # This is our cleaned data to be used for EDA

# to drop duplicate rows
df = df.drop_duplicates()

```

```python
# To reset the index of the dataset
df.reset_index()
```

```python
df.columns
```

```python
# To drop the index column so has not to have repetiitons
df = df.drop(['Unnamed: 0'], axis =1)
```

```python
# to save clean data
df.to_csv('cleaned_data.csv',index = False)
```
